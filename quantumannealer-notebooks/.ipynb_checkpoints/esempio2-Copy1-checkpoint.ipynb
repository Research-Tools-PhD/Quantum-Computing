```python
import numpy as np
from pyqubo import Binary
from pyqubo import Array
import neal
import networkx as nx
import matplotlib.pyplot as plt
import math 
import random
```

```python
def dist(pos,i,j):
    d = (pos[i][0]-pos[j][0])**2+(pos[i][1]-pos[j][1])**2
    return math.sqrt(d)
```

```python
N=20

GR=nx.Graph();
for i in range(N):
    GR.add_node(i)
pos=nx.random_layout(GR)
nx.draw(GR,pos=pos,with_labels = True)
ax = plt.gca()
A = []
for i in range(N):
    center = pos[i]
    A.append(random.uniform(0.1,0.3))
    c = plt.Circle(center, A[i])
    ax.add_patch(c)
#print(A)
plt.axis('equal')
plt.show()
```

```python
B = np.zeros((N,N), dtype=float)
for i in range(N):
    for j in range(i+1,N):
        appo = A[i]+A[j]-dist(pos,i,j)
        if(appo > 0):
            B[i][j] = appo
#print(B)
```

```python
GRe=nx.Graph();
for i in range(N):
    GRe.add_node(i)
for i in range(N):
    for j in range(i+1,N):
        if(B[i][j] > 0):
            GRe.add_edge(i,j)
nx.draw(GRe,pos=pos,with_labels = True)
```

```python
q = Array.create('q', shape=N, vartype='BINARY')
H = 0;

alpha = 1

F = 10 #numero fissato di antenne
beta = 1

for i in range(N):
    H = H + (beta*(1-2*F)-A[i]*A[i]*math.pi)*q[i];
    for j in range(i+1,N):
        H = H + (alpha*B[i][j]*B[i][j]*math.pi + 2*beta)*q[i]*q[j];

model = H.compile()
bqm = model.to_bqm()
```

```python
sa = neal.SimulatedAnnealingSampler()
sampleset = sa.sample(bqm, num_reads=10000)
decoded_samples = model.decode_sampleset(sampleset)
best_sample = min(decoded_samples, key=lambda x: x.energy)
print(best_sample.sample);
```

```python
sol=[]
for i in range(N):
    key='q['+str(i)+']';
    sol.append(best_sample.sample[key])
print(sol)
```

```python
GSol =nx.Graph();
pos_sol = []
A_sol = []
for i in range(N):
    if(sol[i] > 0):
        GSol.add_node(i)
        pos_sol.append(pos[i])
        A_sol.append(A[i])
nx.draw(GSol,pos=pos,with_labels = True)
ax = plt.gca()
for i in range(len(pos_sol)):
    center = pos_sol[i]
    c = plt.Circle(center, A_sol[i])
    ax.add_patch(c)
plt.axis('equal')
plt.show()
```
